Question 2 - Extracting Tesla Revenue Data Using Webscraping - 1 Points

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
# URL for Tesla revenue data
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Fetch the webpage content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Locate the revenue table by inspecting the page structure
table = soup.find_all('table')[1]  # Second table on the page contains revenue data

# Extract table rows
rows = table.find_all('tr')

# Parse the data into a list of dictionaries
revenue_data = []
for row in rows[1:]:  # Skip the header row
    cells = row.find_all('td')
    if len(cells) > 1:
        date = cells[0].text.strip()
        revenue = cells[1].text.strip().replace(",", "").replace("$", "")
        revenue_data.append({"Date": date, "Revenue": float(revenue)})

# Create a DataFrame
tesla_revenue_df = pd.DataFrame(revenue_data)
# Display the last five rows of the DataFrame
print(tesla_revenue_df.tail())
```

```python
# URL for GameStop revenue data (update with a reliable source)
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Extract table rows for revenue data
revenue_table = soup.find_all('table')[1]  # Adjust index based on structure
rows = revenue_table.find_all('tr')

# Parse data into a list of dictionaries
revenue_data = []
for row in rows[1:]:
    cells = row.find_all('td')
    if len(cells) > 1:
        date = cells[0].text.strip()
        revenue = cells[1].text.strip().replace(",", "").replace("$", "")
        revenue_data.append({"Date": date, "Revenue": float(revenue)})

# Convert to DataFrame
gamestop_revenue_df = pd.DataFrame(revenue_data)
print(gamestop_revenue_df.head())
```

Question 5 - Tesla Stock and Revenue Dashboard - 2 Points


```python
import matplotlib.pyplot as plt

# Merge Tesla stock data and revenue data
tesla_data.reset_index(inplace=True)
tesla_merged = pd.merge(tesla_data, tesla_revenue_df, left_on='Date', right_on='Date', how='inner')

# Plot Tesla stock price vs revenue
plt.figure(figsize=(14, 7))
plt.plot(tesla_merged['Date'], tesla_merged['Close'], label='Stock Price', color='blue')
plt.plot(tesla_merged['Date'], tesla_merged['Revenue'], label='Revenue', color='green')
plt.title("Tesla Stock Price vs Revenue")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```

Question 5: Plot Tesla Stock Graph

```python
import yfinance as yf
import matplotlib.pyplot as plt
# Fetch Tesla stock data for the past 6 months
tesla_stock_data = yf.download("TSLA", start="2024-06-01", end="2024-12-01")

# Display the first few rows of the stock data
tesla_stock_data.head()
def make_graph(stock_data, title):
    # Plot the 'Close' price over time
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data.index, stock_data['Close'], label="Tesla Stock Close Price", color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Call the function with Tesla stock data and a title
make_graph(tesla_stock_data, "Tesla Stock Price (TSLA) - June 2024 to December 2024")
```
